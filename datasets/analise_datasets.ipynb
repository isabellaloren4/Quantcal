# Análise dos datasets
- Análise da classe minoritária
- Número de Features 
- Número de instâncias

```python
import os
import pandas as pd

# Caminho da pasta
dataset_path = "datasets"

# Listar apenas arquivos CSV
csv_files = [f for f in os.listdir(dataset_path) if f.endswith(".csv")]

print(f"Datasets encontrados: {csv_files}\n")

for file in csv_files:
    print("=" * 60)
    print(f"Dataset: {file}")

    # Carregar o dataset
    df = pd.read_csv(os.path.join(dataset_path, file))

    # Número de instâncias e features
    num_instancias = df.shape[0]
    num_features = df.shape[1] - 1  # assumindo última coluna como classe

    # Coluna de classe
    target_col = df.columns[-1]

    # Distribuição das classes
    class_counts = df[target_col].value_counts()

    # Classe(s) minoritária(s)
    min_count = class_counts.min()
    minority_classes = class_counts[class_counts == min_count]

    print(f"Número de instâncias: {num_instancias}")
    print(f"Número de features: {num_features}\n")

    print("Distribuição das classes:")
    print(class_counts, "\n")

    print("Classe(s) minoritária(s):")
    print(minority_classes)
```
